## ADDRESS COLUMN

```python
df['City'] = df['ADDRESS'].apply(lambda x: x.split(',')[-1].strip())
df_test['City'] = df_test['ADDRESS'].apply(lambda x: x.split(',')[-1].strip())
```

```python
# Top 10 most frequent cities
print(df['City'].value_counts().head(10))
```

```python
# 1. Extract top cities from train only (threshold = 900)
top_cities = df['City'].value_counts()[df['City'].value_counts() >= 900].index

# 2. Train set processing
df['City_mod'] = df['City'].where(df['City'].isin(top_cities), 'Other')
df_city_dummies = pd.get_dummies(df['City_mod'], drop_first=False).astype(int)
df = pd.concat([df.drop(columns=['City']), df_city_dummies], axis=1)

# 3. Test set processing (use SAME top_cities)
df_test['City_mod'] = df_test['City'].where(df_test['City'].isin(top_cities), 'Other')
df_test_city_dummies = pd.get_dummies(df_test['City_mod'], drop_first=False).astype(int)

# 4. Align test dummies with train dummies (to ensure same columns)
df_test_city_dummies = df_test_city_dummies.reindex(columns=df_city_dummies.columns, fill_value=0)
df_test = pd.concat([df_test.drop(columns=['City']), df_test_city_dummies], axis=1)

print("Train shape:", df.shape)
print("Test shape:", df_test.shape)
```

## One Hot Encoding



```python
# Train
dummies_train = pd.get_dummies(df['POSTED_BY'], drop_first=False).astype(int)
dummies_train = dummies_train.reindex(columns=['Builder', 'Dealer', 'Owner'], fill_value=0)
df = pd.concat([df.drop(columns=['POSTED_BY']), dummies_train], axis=1)

# Test
dummies_test = pd.get_dummies(df_test['POSTED_BY'], drop_first=False).astype(int)
dummies_test = dummies_test.reindex(columns=['Builder', 'Dealer', 'Owner'], fill_value=0)
df_test = pd.concat([df_test.drop(columns=['POSTED_BY']), dummies_test], axis=1)
```

```python
df_test = df_test.drop(columns=['BHK_OR_RK'])
df = df.drop(columns=['BHK_OR_RK'])
```

```python
df['BHK_NO.'] = df['BHK_NO.'].astype(int)
df_test['BHK_NO.'] = df_test['BHK_NO.'].astype(int)
```

```python
df.drop(columns=['ADDRESS'],inplace=True)
df_test.drop(columns=['ADDRESS'],inplace=True)
df.drop(columns=['City_mod'],inplace=True)
df_test.drop(columns=['City_mod'],inplace=True)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(columns=['TARGET(PRICE_IN_LACS)']),df['TARGET(PRICE_IN_LACS)'],test_size=0.2,random_state=42)
```

```python
SS=StandardScaler()
X_train=SS.fit_transform(X_train)
X_test=SS.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

# Initialize model
gbr = GradientBoostingRegressor(
    n_estimators=200,   # number of boosting stages
    learning_rate=0.1,  # step size shrinkage
    max_depth=3,        # depth of individual trees
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

# Train
gbr.fit(X_train, Y_train)

# Predict
y_pred_gbr = gbr.predict(X_test)

# Evaluate
print("Gradient Boosting R Score:", r2_score(Y_test, y_pred_gbr))
```

```python
import joblib
joblib.dump(gbr, 'gradient_boosting_model.pkl')

# Load the model later
loaded_model = joblib.load('gradient_boosting_model.pkl')

# Test prediction
y_pred = loaded_model.predict(df_test)
print(y_pred[:5])
```
